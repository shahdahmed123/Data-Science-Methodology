```python

plt.figure(figsize = (20, 7))
df=df.groupby('Store')['Weekly_Sales'].max().to_frame().reset_index()
#Creating the bar plot
sns.barplot(x = 'Store',y = 'Weekly_Sales',data = df,palette = "Greens"  )
#Adding the aesthetics

plt.title('Bar Chart Stors vs. Max_sales')
plt.xlabel('Store_number')
plt.ylabel('max_sales')
#Show the Bar Chart Plot
plt.show()

```

```python
#e) Plot the relations between weekly sales vs. other numeric features and give insights.




plt.figure(figsize = (50, 7))
df=df.groupby('Weekly_Sales')['Temperature'].mean().to_frame().reset_index()
#Creating the bar plot
sns.barplot(y = 'Temperature',x = 'Weekly_Sales',data = df,palette = "Blues")
#Adding the aesthetics Column chart
plt.title('Bar Chart Temperature vs. Weekly sales')
plt.xlabel('mean Temperature')
plt.ylabel('weekly sales')
#Show the Bar Chart Plot
plt.show()
```
