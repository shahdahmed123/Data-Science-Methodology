```python
#visualization
#Weekly_Sales

# Histogram for Weekly_Sales
plt.figure(figsize = (10, 7))
x = df["Weekly_Sales"]
plt.hist(x, bins = 20, color = "green")
plt.title("Weekly_Sales")
plt.xlabel("Weekly_Sales")
plt.ylabel("count")
plt.show()



# Histogram for Temperature
plt.figure(figsize = (10, 7))
x = df["Temperature"]
plt.hist(x, bins = 20, color = "pink")
plt.title("Temperature")
plt.xlabel("Temperature")
plt.ylabel("count")
plt.show()


# In[142]:


# Histogram for Fuel_Price
plt.figure(figsize = (10, 7))
x = df["Fuel_Price"]
plt.hist(x, bins = 20, color = "red")
plt.title("Fuel_Price")
plt.xlabel("Fuel_Price")
plt.ylabel("count")
plt.show()

# Histogram for CPI
plt.figure(figsize = (10, 7))
x = df["CPI"]
plt.hist(x, bins = 20, color = "yellow")
plt.title("CPI")
plt.xlabel("CPI")
plt.ylabel("count")
plt.show()


# In[148]:


# Histogram for Unemployment
plt.figure(figsize = (10, 7))
x = df["Unemployment"]
plt.hist(x, bins = 20, color = "purple")
plt.title("Unemployment")
plt.xlabel("Unemployment")
plt.ylabel("count")
plt.show()

```
